## 1. Introduction

**Motivation** : A big data volume implies a need for *scalable* and *performance-concsious* analytics algorithms.
Our MapReduce implementation of the clustering algorithm AP includs: 
 - a parallelization stategy of the examplar-based clustering algorithm AP 
 - an improvement of the complexity from a quadratic run-time to a linear one
 - an extension to multilevel Hierachical AP
 - a validation on synthetic and real datasets



**MapReduce** is one of the leading parallelization strategies. Introduced by Google, it is a programming paradigm designed to facilitate the distribution of computations on a cluster of computers. The model is made of two phases : 


 1. The **mapper phase** using key value identifiers for the distribution of data and subsequent independant execution to generate intermediate results
 2. The **reducing phase** that gathers intermediate results to produce the final output key-value pairing.
 
 
In our case, we want to implement the method for a clustering algorithm. 

**Clustering Techniques** are unsupervised solutions to aggregate similar data paterns. It could be the key to discover meaningful insightsand latent trends. Some clustering solutions like the K-means or Gaussian Mixture Models require the *number of cluster centers* which biasedly group the data. The algorithM of **Affinity Propagation** (AP) has an exemplar-based clustering approach. It does not seek to find a mean for each cluster center but instead select representative data points as exemplar of clustered sub-groups, it does not require pre-set K cluster numbers. It is both based on a **"message passing""** framework where data points "talk" to each other in order to determine the most likely exemplar and autimatically determine the clusters and relies on a **pairwise similarity measure** between all data points. The extension of AP,
**The Hierachical Affinity Propagation** (HAP) is an extension of AP clustering which groups and stacks data in a tiered manner. It only required the number of hierachy levels as input but has a limited usage due to quadratic complexity. This is why a MapReduce implementation for HAP is an effective parallelization strategy to solve the curse of run-time complexity.

## 2. The Affinity Propagation Algorithm

The algorithm AP relies on measuring some properties shared by two points of a given dataset. To complete the clustering process, we need to define :  

- **The similarity s** between two points :
    - $s(i,k) = -||x_i - x_k||^2$ for $i\neq k$ , negative squared error with euclidean distance which measures how the data point $x_k$ is accurate to be the center of data point $x_i$
    - $s(k,k) = median(s(i,k))$ are called "preferences". It represents how likely a data point $x_k$ want to choose itself as the cluster center. These values are set to a common (changeable) value. In our case, it will be set to the median of the input similarities. That lead to a moderate number of clusters. Choosing the minimum would lead to a small number of clusters.


- **The availability a** and the **responsibility r** between two points :
   
    - $a(i,k)$ are sent from point i to point k. It shows how accurate is for point i to choose point k as their cluster center. (For t=0, it is set to 0.)
   
    - $r(i,k)$ : sent from point i to point k shows how well-suited point k is to be served as the cluster center for point i. 
    
The responsibility is computed by : 
 
 $$ r(i,k) \leftarrow s(i,k) - max_{k'|k'\ne k} \{a(i,k') + s(i,k')\} $$
 
Then, the availability gathers evidence from data points to know if each candidate center would make a good cluster center : 
 
 $$ a(i,k) \leftarrow \min \{ 0, r(k,k) + \sum_{i'|i'\notin\{i,k\}} \max \{ 0,r(i',k)\}\}$$
 
 Although $r(k,k)$ is updated as all other pair of points, $a(k,k)$ is updated separatly :
 
 $$ a(k,k) \leftarrow \sum_{i'|i'\ne k} \max \{0,r(i',k)\}$$ 
 
 
Additionnaly, we introduce some settings such as :
 - **The damping step** : in order to avoid numerical oscillations when updating the messages, each message is "dampen" by :
 
 $$ updatedMessage \leftarrow updatedValue \times (1 - \lambda) + oldValue \times \lambda $$
 
 Usually, $\lambda$ is set to 0.5.
 
 - **The three terminal conditions** : 
     1. fixed number of iterations
     2. changes of messages below a threshold value
     3. local decision stays constant for some number of iterations.
     
 - **The attribution of cluster center** :
 For each point i, the point k that maximises $a(i,k) + r(i,k)$ identify the data center for point i. (It might be i itself if $k=i$.) 
 
 - **Iteration step** : 
    1. updating all responsibilities given the availabilities
    2. updating availabilities given the responsibilities
    3. combining availabilities and responsibilities to monitor the center decisions and to decide termination or not.

## 3. Parallelization 

$\rightarrow$ MapReduce environment

#### Map Phase 

- We need to randomly divide the input data into small pieces and store both the raw input data and the divided input data in HDFS. The different steps of the phase are constructed as followed :
    1. Initialize s(k,k) to zero : $s(k,k) = 0$
    2. Compute maximal preferance $P_{max} = \max \{s(i,k)\}$
    3. Compute the minimal preference :
$ dpsim1 = \max_j \{ \sum_j s(i,j) \} $ and 
$ dpsim2 = \max_{i \ne j} \{ \sum_k \max \{ s(i,k),s(j,k)\} \}$
    4. Compute the minimal preference : $P_{min} = dpsim1 - dpsim2 $


- Each mapper will preform the AP algorithm on their own piece of data. Each mapper has its own key and the output returned is a data tuple $<key,i,k>$.

Remarks : 
We could wonder why each mapper should perform the AP independently on its own data.
1. The AP algorithm needs about hundreds of iterations to perform. If all the information is exchanged at each iteration, a huge communication cost is expected in the MapReduce environment. 
2. Every Mapper is isolated n processing which is restricted by the architecture to transfer data among mappers.

#### Reduce Phase 

The reduce phase consists in 2 parts : 
1. **classify the points** : Reducer A classifies the clustered points and sends the results to Reducer B which calculate the centers of clusters. If distance between two centers is less than a given treshold, both clusters are combined into one. In practise, we use $threshold = preference*0.5$. Points which are far away from centers are classified as *isolated points*. Points in the same clusers are sent to Reducer B with the same key.
2. **calculate the center points** : if there are multiple centers in a clusters we calculated the $centroids = \frac{c_1 + c2 + ... + c_K}{K}$.

### Small Test
- We implemented a small test on generated data points

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

#Define 3 centers 
centers = [(1,1), (5,5), (1,5)]

#Generate the datapoints randomly
def Generator(n, centers):
    while n>0 :
        center = centers[np.random.choice(len(centers))]
        yield center[0] + np.random.rand(1)[0]*2,  center[1] + np.random.rand(1)[0]*2
        n -= 1
        
points = [coord for coord in Generator(500, centers)]

#Fit the MapReduce AP on the generated points
MPAP = MapReduceAffinityPropagation(num_workers=2)
MPAP.fit(np.array(points), 2)

#plot the input data points and the resulting clusters
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.scatter([x for x,y in points], [y for x,y in points])
plt.title('Original Points')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.scatter([x for x,y in points], [y for x,y in points], c = MPAP.predict(np.array(points)))
plt.title('Clustered Points')
plt.xlabel('x')
plt.ylabel('y')
```

### Run Time Comparaison
- We plot the running-time complexity against the number of observations, for the AP algorithms and MapReduce implementations for different core values used in the parallelization. To do so, we use a generated dataset.
- The results show that the more we have observations, the more efficient is the MapReduce process  

```python
#from sklearn.cluster import AffinityPropagation
import time
import gc
gc.collect()

run_time_dictionnary = {0 : [], 2 : [], 4 : [],6 : []}

algorithme = {0: AffinityPropagation(), 
              2 : MapReduceAffinityPropagation(num_workers=2),
              4 : MapReduceAffinityPropagation(num_workers=4),
              6 : MapReduceAffinityPropagation(num_workers=6)}

    
centers = [(1,1), (5,5), (1,5)]

def Generator(n, centers):
    while n>0 :
        center = centers[np.random.choice(len(centers))]
        yield center[0] + np.random.rand(1)[0]*2,  center[1] + np.random.rand(1)[0]*2
        n -= 1

init_point = [coord for coord in Generator(30, centers)]
for cores in [0,2,4,6] : 
    start = time.time()
    a = algorithme[cores]
    a.fit(init_point)
    end = time.time() - start
    print('Load Time with {} cores :'.format(cores), end )

for it in [50, 100, 500, 1000, 2000, 3000]:

    points = [coord for coord in Generator(it, centers)]
    
    for cores in [0,2,4,6]:
        start = time.time()
        ap = algorithme[cores]
        ap.fit(np.array(points))
        #ap.predict(points)
        end = time.time() - start

        l = run_time_dictionnary[cores]
        l.append(end)
    
    del points

#plt.scatter([x for x,y in points], [y for x,y in points], c = MRAP.predict(np.array(points)))

plt.figure(figsize = (15,5))

for cores in [0, 2 , 4, 6]:
    plt.plot([50, 100, 500, 1000, 2000, 3000], run_time_dictionnary[cores])

plt.legend(['AP', 'MapReduce AP - 2 Cores', 'MapReduce AP - 4 Cores ', 'MapReduce AP - 6 Cores' ])
plt.title('Comparaison of running Time between AP and MPAP with variable nb of cores')
plt.show()
```

The above comparaison was not realized with the true running time for each algorithm, or not really. Indeed, we decided to load the model and to "activate" the multiprocessing call before running the algorithm with the data. What we want to show here, is that it takes a not negligible time to be loaded properly but with this time apart, we show that the map reduce version is clearly faster, all the more fatser as it has more cores.

Below, you will find the *true* run time of each algorithm if you had to load and train it at the same moment.

```python
import time

ap_time = []
mrap_time = []
mrap_time4 = []
mrap_time8 = []
for it in [50, 100, 500, 1000, 2000]:
    
    centers = [(1,1), (5,5), (1,5), (10,10),(15,15),(10,0), (15,5)]

    def Generator(n, centers):
        while n>0 :
            center = centers[np.random.choice(len(centers))]
            yield center[0] + np.random.rand(1)[0]*2,  center[1] + np.random.rand(1)[0]*2
            n -= 1

    points = [coord for coord in Generator(it, centers)]
    
    start = time.time()
    ap = AffinityPropagation()
    ap.fit(points)
    ap.predict(points)
    end = time.time() - start
    ap_time.append(end)
    
    start = time.time()
    MRAP = MapReduceAffinityPropagation(num_workers=2)
    MRAP.fit(np.array(points))
    end = time.time() - start
    mrap_time.append(end)
    
    start = time.time()
    MRAP = MapReduceAffinityPropagation(num_workers=4)
    MRAP.fit(np.array(points))
    end = time.time() - start
    mrap_time4.append(end)
    
    start = time.time()
    MRAP = MapReduceAffinityPropagation(num_workers=6)
    MRAP.fit(np.array(points))
    end = time.time() - start
    mrap_time8.append(end)
```

```python
plt.figure(figsize = (15,5))

for algo in [ap_time,mrap_time,mrap_time4,mrap_time8]:
    plt.plot([50, 100, 500, 1000, 2000], algo)

plt.legend(['AP', 'MapReduce AP - 2 Cores', 'MapReduce AP - 4 Cores ', 'MapReduce AP - 6 Cores' ])
plt.title('Comparaison of running Time between AP and MPAP with variable nb of cores')
plt.show()
```

The main conclusion of this graph is that it is not *always* interesting to used the map reduce version of the affinity propagation. Indeed, if you have a number of samples smaller than (approximately) 1000 data points, it is faster to use the standard version, because of the loading time (between 2 and 4 seconds). But, and this all the point of the map reduce version, if you use more points, the algorithm becomes easily scalable with the map reduce implementation and runs as fast as the standard AP with $\frac{\mbox{n_samples}}{k}$
